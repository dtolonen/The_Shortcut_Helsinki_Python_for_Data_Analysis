```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('cars.csv')

# remove dollar sign and convert text to number
cars.MSRP = cars.MSRP.str.replace('$', '')
cars.MSRP = cars.MSRP.astype(int)

cars.columns = cars.columns.str.replace(' ', '') # remove white space in column name

cars.head(3)
```

### Groupby

```python
# Son's solution: 
# cars.groupby('Origin').mean()



# Group the data by Origin and calculate the mean for all columns

gb = cars.groupby('Origin') # collapses all the lines into fewer lines by Origin
gb # shows the content of gb? Commenting this line out returns the same
gb.mean().round(2) # what is the figure in the brackets? See documentation!
```

```python
# Son's solution: 
# # Group the data by Make and then by Type. What's the cheapest price for a Volvo's SUV?
# cars.groupby(['Make', 'Type']).min()
# 
# 
# # Alternatively
# # gb = cars.groupby(['Make', 'Type']).min()
# # gb.loc[('Volvo', 'SUV'), 'MSRP']







# Group the data by Make and then by Type. What's the cheapest price for a Volvo's SUV?

# Found this in Lecture 7, under Groupby multi level
min_price = cars.groupby(['Make', 'Type']).MSRP.min()

min_price
```

```python
# son's solution: 
# cars.groupby(cars.Make == 'Audi').mean()



# Are Audi cars heavier than the rest of car brands on average?

# Found this in Lecture 7, under Group by condition
gb = cars.groupby(cars.MSRP == 'Audi')

gb.mean()
```
