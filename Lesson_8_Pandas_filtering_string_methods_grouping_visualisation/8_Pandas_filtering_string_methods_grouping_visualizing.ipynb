You can apply multiple functions on one column, and apply different functions on different columns using `.agg()`

```python
gb.agg({'MSRP': ['max','min'], 'MPG_City': 'mean'})
```

13 agg available functions

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values


#### Custom function for agg()

!!! **For advanced users if you want to see really obscure statistics:** You can **write a custom function** for `agg()` or `apply()`. Check out `apply()` method in the documentation. 

(you'll probably never use these, but **Son wants to show you these** to show how flexible Pandas is - if you have a problem you can create your own functions, which will be faster than writing a for loop, even though your functions will probably not be as optimised as the stock functions.)

The difference between `agg()` and `apply()` is that in `agg()`, you're passing each column at a time into your function. In `apply()`, you're passing the entire dataframe, so your function can perform calculation using multiple columns as input.

```python
#This function calculates the percentage of a certain value for each group
def pct(series, query):
    return series[series == query].count() / len(series)

gb.Type.agg(pct, query='SUV') # calculates the percentage of SUV cars for each car brand group
```

#### Pandas plot() method

Pandas has a convenient `plot()` wrapper for matplotlib. Since it calls matplotlib upon execution, you can include parameters that exists in the corresponding matplotlib function. This is why you should import matplotlib too, as it uses it and calls it (a matplotlib function?) in the background.

For example: check out the documentation on bar plot in matplotlib: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html

You can include parameters such as `color=` in your pandas `plot()` call, which is documented in the `plt.bar()` function.

```python
car_type = cars.Type.value_counts().plot(kind='bar', rot=30, color='g')

# kind - here, you want a bar chart. You can specify all kinds of parameters.
#Check pandas.plot method on pandas.pydata.org site, you'll see all the parameters.

# Arda:: how do you get this to export ?
# Son checks the matplotlib.org site for a function: matplotlib.pyplot.savefig
```

_This is how it's done from scratch using matplotlib_

```python
# this is what you have to do to get the same chart 
# using matplotlib from scratch. So, it's much easier the other way.

t = cars.Type.value_counts()
t_x = np.arange(len(t.index))
t_y = t.values

plt.figure()

plt.bar(t_x, t_y, color='g')

plt.xticks(t_x, t.index.values, rotation=30)
```

!!! If you want to plot multiple different charts in one canvas (figure), you have to use `subplots`.

```python
# Create a figure containing 1 row and 2 columns,  on which you can plot 2 graphs (2 axes)
# So, you need (to plot?) a subplot
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6)) # cols:: these become the two images

# When you assign a plot to a variable (mpg_city), that variable can be used as an ax.
mpg_city = cars.MPG_City.value_counts().sort_index().plot(kind='line', color='black', ax=ax[0])

mpg_highway = cars.MPG_Highway.value_counts().sort_index().plot(kind='line', color='blue', ax=ax[1])
```
