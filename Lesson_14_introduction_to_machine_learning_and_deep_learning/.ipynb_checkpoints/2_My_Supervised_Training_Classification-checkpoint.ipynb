```python
# Creating the feature X and target Y
X = data_frame.drop(["life", "Region"], axis=1)
y = data_frame.loc[:,"life"]
```

```python
print("Shape of X:", X.shape)
print("Shape of Y:", y.shape)
```

```python
# Splitting the data to train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", y_train.shape)
print("Shape of Y_test:", y_test.shape)
```

```python
# Create the regression algorithm
from sklearn.linear_model import LinearRegression
reg_lin = LinearRegression()
reg_lin
```

```python
# Fitting the regressor to the training data
reg_lin.fit(X_train, y_train)
```

```python
# Predicting the unseen data(X_test)
y_pred = reg_lin.predict(X_test)
```

```python
y_pred
```

```python
# Finding the accuracy of the data
print("R2:{}".format(reg_lin.score(X_test,y_test)))
```

```python
# Create a linear regressor
# this is just a name - cv for cross validation, 4 for 4 fold
reg_4_cv = LinearRegression()
reg_4_cv # this is a regressor
```

```python
# Compute 4_fold cross validation scores
from sklearn.model_selection import cross_val_score
# Again, these are just names, in the standard way to do them. 
# and this is just Ali's way to do a script. You can do it your own way. 
cv_scores_4 = cross_val_score(reg_4_cv, X, y, cv=4) 
# ie. "I want the cross validation to take 4 parts/divisions"
cv_scores_4
```

```python
# All good, but the thing is this is not the final accuracy 
# of our functino here: we should get the average. 
# The final value in cv is the average of those divisions.

print("Average 4-fold CV score: {}".format(np.mean(cv_scores_4)))
```

```python
# So, what this just did: 
# it divided the whole data into parts 
# and took one part/division as test 
# and held the other parts as training
```

#### K-Fold CV comparison
Cross validation is essential but do not forget that the more folds you use, the more `computationally expensive` cross-validation becomes. Here we will compare `3-fold` cross-validation and `10-fold` cross-validation on the Gapminder dataset.

```python
# K-Fold CV comparison

# In Machine Learning, two key words are very important: 
# overfitting and underfitting
# if the data is very complex, the model will (hang?), 
# but if the model is very simple (we won't get anything)
# So, now, we're doing all the basics, so that we can get
# to the real work - optimisation. 

# Like here, we don't have too much data, like 200 (samples?), 
# but think if you have like 2 million - then the computation
# will get very expensive, so this (model?) won't be useful to use.
```

```python
# Create a linear regression
reg_cv = LinearRegression()

# 3-Folds Cross Validation
cv_scores_3 = cross_val_score(reg_cv, X, y, cv=3) 
print("Average 3-Fold CV score: {}".format(np.mean(cv_scores_3)))
# we use this function from above, and what are the inputs, it's our regressor.

# 10-Folds Cross Validation
cv_scores_10 = cross_val_score(reg_cv, X, y, cv=10)
print("Average 10-Fold CV score: {}".format(np.mean(cv_scores_10)))
```

```python
# Import Lasso
from sklearn.linear_model import Lasso

# and this below Lasso is just a name
Lasso = Lasso(alpha=0.4, normalize=True)
Lasso
```

```python
# Fir the regression

Lasso.fit(X,y)
```

```python
print("R2: {}".format(Lasso.score(X,y)))
```

```python
# Compute the coefficients
lasso_coef = Lasso.coef_
print(lasso_coef)
```

```python
x_features = X.columns
x_features
```

```python
# plotting the coefficients
# you can use lasso function/regressor to find what are 
# the most important out of your 8 features 
# (ie. all your column headings that are not numerical 
# (our 'target') but text)

plt.plot(range(len(x_features)), lasso_coef)
# here, we should be careful with the lowercase l in lasso etc.
plt.xticks(range(len(data_frame.columns)),x_features.values, rotation=60)
plt.margins(0.02)
plt.show() 
```

According to the lasso algorithm, it seems like `child_mortality` is the most important feature when predicting `life expectancy`.

```python
# all of these are zero, but for child mortality it's bigger than zero. 
# This means that child mortality is ths most weighted, or most important
# feature (in predicting?) life expectancy.

# And - We can go up to our heat map to see what it tells us. yes, it's
# among the biggest ones. 

# But Lasso is more trustworthy than heat map. 

# Whenever you have a regression problem, first run the Lasso function,
# then you will find the most weighted feature, then you can do your 
# regression.

# So this is one way to look at it. 
```

```python
y_life = data_frame['life'].values
X_fertility = data_frame['fertility'].values
X_mortality = data_frame['child_mortality'].values
X_HIV = data_frame['HIV'].values

print("Dimensions of y_life before reshaping: {}".format(y_life.shape))
print("Dimensions of X_fertility before reshaping: {}".format(X_fertility.shape))
print("Dimensions of X_mortality before reshaping: {}".format(X_mortality.shape))
print("Dimensions of X_HIV before reshaping: {}".format(X_HIV.shape))
```

```python
y_life = y_life.reshape(-1,1)
X_fertility = X_fertility.reshape(-1,1)
X_mortality = X_mortality.reshape(-1,1)
X_HIV = X_HIV.reshape(-1,1)

print("Dimensions of y_life after reshaping: {}".format(y_life.shape))
print("Dimensions of X_fertility after reshaping: {}".format(X_fertility.shape))
print("Dimensions of X_mortality after reshaping: {}".format(X_mortality.shape))
print("Dimensions of X_HIV after reshaping: {}".format(X_HIV.shape))
```

```python
# Life expectancy vs Fertility
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

prediction_space = np.linspace(min(X_fertility),max(X_fertility)).reshape(-1,1)

reg.fit(X_fertility,y_life)

y_pred = reg.predict(prediction_space)

# Print R^2
print(reg.score(X_fertility,y_life))

# Plot regression line
plt.scatter(X_fertility,y_life,color='green',linewidth=3)
plt.plot(prediction_space,y_pred,color='black',linewidth=3)
plt.show()
```

```python
# Life expectancy vs child mortality
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

prediction_space = np.linspace(min(X_mortality),max(X_mortality)).reshape(-1,1)

reg.fit(X_mortality,y_life)

y_pred = reg.predict(prediction_space)

# Print R^2
print(reg.score(X_mortality,y_life))

# Plot regression line
plt.scatter(X_mortality,y_life,color='green',linewidth=3)
plt.plot(prediction_space,y_pred,color='black',linewidth=3)
plt.show()
```

```python
# Life expectancy vs HIV
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

prediction_space = np.linspace(min(X_HIV),max(X_HIV)).reshape(-1,1)

reg.fit(X_HIV,y_life)

y_pred = reg.predict(prediction_space)

# Print R^2
print(reg.score(X_HIV,y_life))

# Plot regression line
plt.scatter(X_HIV,y_life,color='green',linewidth=3)
plt.plot(prediction_space,y_pred,color='black',linewidth=3)
plt.show()
```
