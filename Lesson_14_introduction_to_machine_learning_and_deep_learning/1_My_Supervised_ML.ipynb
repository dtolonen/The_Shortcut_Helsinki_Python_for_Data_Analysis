#### K-Nearest Neighbors (KNN)

* KNN: Look at the K closest labeled data points

* Classification method

* First we need to train our data. (Train=fit)

* fit(): fits the data, train the data

* predict(): predicts the data

* Lets learn how to implement it with sklearn:
  - x: features
  - y: target variables(Normal,Abnormal)
  - n_neighbors: k. In this example it is 3. it means that look at the 3 closest labeled data points

```python
# Implement the KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn
```

```python
# Creating the feature=x and target=y
x = data_frame.loc[:, data_frame.columns!="class"]
y = data_frame.loc[:,"class"]
```

```python
print("Shape of x:",x.shape)
print("Shape of y:",y.shape)
```

```python
# Fitting the data to KNN algorithm - here, it's learned 
knn.fit(x,y)
```

```python
# Prediction - 
prediction_knn = knn.predict(x)
prediction_knn
```

```python
print("with knn(k=3) amd (x,y) accuracy is:" , knn.score(x,y))

# Ali: Fabio pointed out a good point: what 
# we created out classifier
# introed out features and labels
# we just showed the classifier every dataset, 
# so we trained the machine with the entire data
# so it'll be very easy for the machine to guess the labels
# but in the real world we'll want to put data into the machine
# that the machine has not yet seen.
# so, we split our data to train and test the (new) data
# so for that 
```

```python
# Splitting the data to train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Fitting
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
# Ali: Now, let's go to the prediction
# Prediction

# Ali: so, below, a prediction of knn, which has been spli
prediction_knn_split = knn.predict(x_test)
```

```python
print("with knn(k=3) amd (x,y) accuracy is:" , knn.score(x,y))
```

### B. Regression

#### LinearRegression

Now we have our data to make regression. In `regression` problems `target` value is `continuously varying variable` such as price of house or sacral_slope. Lets `fit line into this points`.

* `y = ax+b` where `y=target`, `x=feature` and `a=parameter` of model

* We choose parameter of model(a) according to minimum error function that is `lost function`

* In linear regression we use `Ordinary Least Square (OLS)` as lost function.

* `OLS`: sum all residuals but some positive and negative residuals can cancel each other so we sum of square of residuals. It is called OLS

* `Score`: Score uses `R^2` method that is `((y_pred-y)^2)/(y_actual-y)^2`

```python
# Create Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg
```

```python
# Predict Space
predict_space = np.linspace(min(X),max(X)).reshape(-1,1)
predict_space
```

```python
# Fitting
reg.fit(X,y)
```

```python
# Prediction
y_predict = reg.predict(predict_space)
```

```python
print("Accuracy:",reg.score(X,y))
```

```python
# Plotting
plt.scatter(x=X,y=y)
plt.plot(predict_space,y_predict,color="black",linewidth=3)
plt.xlabel("pelvic_incidence")
plt.ylabel("scral_slope")
plt.show()
```
