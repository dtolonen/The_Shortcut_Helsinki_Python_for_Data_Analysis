(Optional) Write a function that returns all the column pairs with a correlation higher than 0.7 or lower than -0.7

Example output: 

- Wheelbase and Length: 0.89
- MPG_Highway and Weight: -0.79

```python
corr = cars.corr()

corr
```

```python
''' Notice that the corr table is symmetrical, separated by the diagonal values. 
Therefore we only need to check the top half or bottom half of the table.
Here I'm choosing to check only the top half.
By iterating j from i+1, I can avoid checking the diagonal values and the bottom half.

Also, I mentioned that you should avoid using for-loop to loop through your data.
However, here it is acceptable because the correlation table is usually very small.
You are not looping through hundreds or thousands of row. So there is no performance issue here.
'''

def high_corr(corr, threshold):
    pair_list = []
    
    for i in range(0,len(corr)):
        for j in range(i+1,len(corr)):
            if (corr.iloc[i,j] > threshold) or (corr.iloc[i,j] < -threshold):
                pair_list.append((corr.iloc[i,j], i, j))

    return pair_list
    #If you want to show higher ones first            
    #return sorted(pair_list,key=lambda x: -abs(x[0]))


cols = corr.columns
for val, i, j in high_corr(corr, 0.7):
    print(f'{cols[i]} and {cols[j]}: {val:.2f}')
```
